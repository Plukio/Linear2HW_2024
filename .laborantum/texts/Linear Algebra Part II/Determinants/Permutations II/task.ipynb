```python
from itertools import permutations

def permutation_sign(perm):
    """Calculate the sign of a permutation (+1 or -1)."""
    inversions = 0
    for i in range(len(perm)):
        for j in range(i + 1, len(perm)):
            if perm[i] > perm[j]:
                inversions += 1
    return -1 if inversions % 2 else 1

def determinant(A):
    # Ensure the matrix is square
    n = A.shape[0]
    assert A.shape[0] == A.shape[1]

    det = 0
    elements = []

    ### YOUR CODE HERE ###
    for perm in permutations(range(n)):
        sign = permutation_sign(perm)
        product = 1

        for i in range(n):
            product *= A[i, perm[i]]

        det += sign * product
        elements.append((sign, perm, product))

    return det, sorted(elements)
```

```python
inputs = json_tricks.load('inputs.json')
results = {'results': []}

for args in inputs['inputs']:
    results['results'].append([determinant(**args)])

json_tricks.dump(results, '.answer.json')
```
