```python
def reconstruct(A, b):
    U, s, VT = np.linalg.svd(A)
    results = []
    ### YOUR CODE HERE ###

    
    n = len(s)


    for k in range(1, n + 1, 16):
        res = np.zeros_like(A[1, :])
        U_k = U[:, :k] 
        s_k = s[:k]
        VT_k =  VT[:k, :]
        for i in range(len(s_k)):
            res += ((U_k[:, i] @ b) / s_k[i]) * VT_k[i,:].T
        results.append(res)
        
    return results, U, s, VT
```

```python
inputs = json_tricks.load('.inputs.json')

# print(inputs)

x_rec, U, s, VT = reconstruct(**inputs)

json_tricks.dump(x_rec, '.answer.json')
```
